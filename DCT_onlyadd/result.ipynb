```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

base_path = "C:/Users/kanta/Desktop/prom/20230910/1_11/mset_"
suffix = "_IDVGS.csv"

dfs = []

# 5.013.0CSV
for i in range(3, 9):
    formatted_number = i
    file_name = f"{base_path}{formatted_number}{suffix}"
    df = pd.read_csv(file_name, usecols=["time", "ig"])
    
    # 500 500
    num_segments = len(df) // 500
    labels = []
    for segment in range(num_segments):
        labels.extend([f"{formatted_number}.{segment}"] * 500)
    remaining = len(df) - num_segments * 500
    labels.extend([f"{formatted_number}.{num_segments}"] * remaining)

    df["source"] = labels
    dfs.append(df)

# 
df_1_7_merge = pd.concat(dfs, axis=0, ignore_index=True)


# 
plt.figure(figsize=(20,8))
sns.lineplot(data=df_1_7_merge, x=df_1_7_merge.index, y="ig", hue="source", palette="tab20")
plt.xlim([0,400*80])
#plt.yscale('log')
plt.ylim([1E-11, 1E-6])
plt.legend(loc="upper center", bbox_to_anchor=(0.5, -0.15), ncol=10)
#plt.tight_layout()
plt.title("1_10, current limit 10mA")
plt.show()

#500
```
